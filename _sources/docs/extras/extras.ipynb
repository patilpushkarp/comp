```python
import pandas as pd
```

## Join word distribution results

```python
fintechs = ['faircent', 'lendingkart', 'mobikwik', 'paytm', 'pine_labs']
nbfcs = ['aditya_birla_finance', 'bajaj_finance', 'muthoot_finance']
```

```python
df_list = []
for fintech in fintechs:
    pre_wd_df = pd.read_csv(f'./../data/fintech/{fintech}/pre_wd.csv')
    pre_wd_df['total_tweets'] = pre_wd_df['true_cnt'] + pre_wd_df['false_cnt']
    post_wd_df = pd.read_csv(f'./../data/fintech/{fintech}/post_wd.csv')
    post_wd_df['total_tweets'] = post_wd_df['true_cnt'] + post_wd_df['false_cnt']
    pre_wd_df.set_index('words', inplace=True)
    post_wd_df.set_index('words', inplace=True)
    result_df = pd.concat([pre_wd_df, post_wd_df], axis=1)
    result_df = result_df['true_cnt']
    result_df.columns = ['Pre-COVID', 'Post-COVID']
    new_row = {"Pre-COVID": pre_wd_df['total_tweets'].unique()[0], "Post-COVID": post_wd_df['total_tweets'].unique()[0]}
    result_df = result_df.append(pd.Series(new_row, index=result_df.columns, name='Total Tweets'))
    cols = pd.MultiIndex.from_tuples([(f"{fintech.title()}", "Pre-COVID"),
                                        (f"{fintech.title()}", "Post-COVID")])
    df = pd.DataFrame(result_df.values.tolist(), columns=cols, index=result_df.index)
    df_list.append(df)

fintech_result = pd.concat(df_list, axis=1)
```
